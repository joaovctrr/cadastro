```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchWindowException
import time
import pandas as pd
import openpyxl
```

```python
dados = pd.read_excel('arquivo.xlsx')
```

```python
# Configurando o WebDriver
driver = webdriver.Chrome()
driver.get('URL DO SITE')
wait = WebDriverWait(driver, 5)

# Esperando pelo elemento de sucesso após a página carregar
sucesso = wait.until(EC.presence_of_element_located((By.XPATH, 'CAMINHO_DO_XPATH')))

def cadastrar_funcionario(numero, digito, nome):
    try:
        # Preenchendo os campos do formulário
        numero_input = wait.until(EC.presence_of_element_located((By.XPATH, 'CAMINHO_DO_XPATH')))
        numero_input.click()
        numero_input.send_keys(f'{numero}{digito}')

        agente_input = wait.until(EC.presence_of_element_located((By.XPATH, 'CAMINHO_DO_XPATH')))
        agente_input.click()
        agente_input.send_keys(nome)

        bhpro_input = wait.until(EC.presence_of_element_located((By.XPATH, 'CAMINHO_DO_XPATH')))
        bhpro_input.click()
        bhpro_input.send_keys(f'{numero}{digito}')

        bhpro_select = wait.until(EC.presence_of_element_located((By.XPATH, 'CAMINHO_DO_XPATH')))
        bhpro_select.click()
        bhpro_select.send_keys('EMPRESA')
        bhpro_select.send_keys(Keys.ENTER)

        botao_salvar = wait.until(EC.presence_of_element_located((By.XPATH, 'CAMINHO_DO_XPATH')))
        botao_salvar.click()       
        
        time.sleep(2)

        # Limpando campos após salvar
        limpar_numero_input = wait.until(EC.presence_of_element_located((By.XPATH, 'CAMINHO_DO_XPATH')))
        limpar_numero_input.click()
        limpar_numero_input.clear()

        limpar_agente_input = wait.until(EC.presence_of_element_located((By.XPATH, 'CAMINHO_DO_XPATH')))
        limpar_agente_input.click()
        limpar_agente_input.clear()

        limpar_bhpro_input = wait.until(EC.presence_of_element_located((By.XPATH, 'CAMINHO_DO_XPATH')))
        limpar_bhpro_input.click()
        limpar_bhpro_input.clear()

        # Tentando lidar com a nova janela para cadastro de senha
        try:
            janelas = driver.window_handles
            driver.switch_to.window(janelas[-1])

            senha_input = wait.until(EC.presence_of_element_located((By.XPATH,  'CAMINHO_DO_XPATH')))
            senha_input.click()
            senha_input.send_keys(f'{numero}{digito}')
            
            confirmar_senha = wait.until(EC.presence_of_element_located((By.XPATH,  'CAMINHO_DO_XPATH')))
            confirmar_senha.click()
            confirmar_senha.send_keys(f'PM{numero}{digito}')
            
            confirmar_clique =  wait.until(EC.presence_of_element_located((By.XPATH, 'CAMINHO_DO_XPATH')))
            confirmar_clique.click()
            print(f'Usuário cadastrado: {nome}  ')
            dados.at[indice, 'REGISTRO'] = '1'

            driver.switch_to.window(janelas[0])

            time.sleep(2)


        except NoSuchWindowException:
            # Se NoSuchWindowException for lançada, significa que a janela não foi encontrada
            # Neste caso, o código pode continuar normalmente
            print('Continuando...')
            
    except Exception as e:
        pass

# Iterando sobre os dados
for indice, linha in dados.iterrows():
    numero = str(linha['NUMERO'])
    digito = str(linha['DV'])
    nome = linha['NOME SERVIDOR']

    print(f'Cadastrando {indice} - {nome} {numero}{digito}')
    feito = cadastrar_funcionario(numero, digito, nome)
    if not feito:
        pass
```
