```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchWindowException
import time
import pandas as pd
import openpyxl
```

```python
dados = pd.read_excel('Arquivo.xslx')
```

```python
driver = webdriver.Chrome()
driver.get('URL')
wait = WebDriverWait(driver, 0.1)

def cadastrar_aparelho(imei, instalacao, fabricante, modelo, nome):
    try:
        # Preenchendo os campos do formul√°rio
        numero_serie_input = (wait.until(EC.presence_of_element_located
                                        ((By.XPATH, 'XPATH_CAMINHO'))))
        numero_serie_input.click()
        numero_serie_input.send_keys(imei)
        time.sleep(0.5)

        codigo_instalacao_input = (wait.until(EC.presence_of_element_located
                                             ((By.XPATH, 'XPATH_CAMINHO'))))
        codigo_instalacao_input.click()
        codigo_instalacao_input.send_keys(instalacao)
        time.sleep(0.5)

        lotacao_agente_input = (wait.until(EC.presence_of_element_located
                                          ((By.XPATH, 'XPATH_CAMINHO'))))
        lotacao_agente_input.click()
        lotacao_agente_input.send_keys('CPPM')
        time.sleep(0.5)

        # controle_patrimonial_input = (wait.until(EC.presence_of_element_located
        #                                         ((By.XPATH, '//*[@id="formulario:dados"]/table/tbody/tr/td/div/table/tbody/tr[6]/td[3]/input'))))
        # controle_patrimonial_input.click()
        # controle_patrimonial_input.send_keys('')
        # time.sleep(0.5)

        fabricante_input = (wait.until(EC.presence_of_element_located
                                      ((By.XPATH, 'XPATH_CAMINHO'))))
        fabricante_input.click()  
        fabricante_input.send_keys(fabricante)
        time.sleep(0.5)

        modelo_input = (wait.until(EC.presence_of_element_located
                                  ((By.XPATH, 'XPATH_CAMINHO'))))
        modelo_input.click()
        modelo_input.send_keys(modelo)
        time.sleep(0.5)

        nome_do_agente = (wait.until(EC.presence_of_element_located
                                           ((By.XPATH, 'XPATH_CAMINHO'))))
        nome_do_agente.click()
        nome_do_agente.send_keys(f'PM {nome}')
        time.sleep(0.5)

        bloqueado_form = (wait.until(EC.presence_of_element_located
                                    ((By.XPATH, 'XPATH_CAMINHO'))))
        bloqueado_form.click()
        bloqueado_form.send_keys('N')
        bloqueado_form.send_keys(Keys.ENTER)
        time.sleep(0.5)

        botao_salvar = (wait.until(EC.presence_of_element_located
                                  ((By.XPATH, 'XPATH_CAMINHO'))))
        botao_salvar.click() 
        print('Cadastrado com sucesso')
        # Salvar: //*[@id="formulario:dados"]/table/tbody/tr/td/div/table/tbody/tr[11]/td[3]/input[1]
        # Limpar: //*[@id="formulario:dados"]/table/tbody/tr/td/div/table/tbody/tr[11]/td[3]/input[2]
        

        # LIMPAR CAMPOS

        limpar_numero_serie_input = (wait.until(EC.presence_of_element_located
                                                ((By.XPATH, 'XPATH_CAMINHO'))))
        limpar_numero_serie_input.click()
        limpar_numero_serie_input.clear()


        limpar_codigo_instalacao_input = (wait.until(EC.presence_of_element_located
                                                     ((By.XPATH, 'XPATH_CAMINHO'))))
        limpar_codigo_instalacao_input.click()
        limpar_codigo_instalacao_input.clear()

        

        limpar_lotacao_do_agente = (wait.until(EC.presence_of_element_located
                                                ((By.XPATH, 'XPATH_CAMINHO'))))
        limpar_lotacao_do_agente.click()
        limpar_lotacao_do_agente.clear()

        # limpar_controle_patrimonial_input = (wait.until(EC.presence_of_element_located
        #                                                 ((By.XPATH, '//*[@id="formulario:dados"]/table/tbody/tr/td/div/table/tbody/tr[6]/td[3]/input'))))
        # limpar_controle_patrimonial_input.click()
        # limpar_controle_patrimonial_input.clear()

        limpar_fabricante_input = (wait.until(EC.presence_of_element_located
                                              ((By.XPATH, 'XPATH_CAMINHO'))))
        limpar_fabricante_input.click()  
        limpar_fabricante_input.clear()

        limpar_modelo_input = (wait.until(EC.presence_of_element_located
                                          ((By.XPATH, 'XPATH_CAMINHO'))))
        limpar_modelo_input.click()
        limpar_modelo_input.clear()

        limpar_nome_do_agente = (wait.until(EC.presence_of_element_located
                                                   ((By.XPATH, 'XPATH_CAMINHO'))))
        limpar_nome_do_agente.click()
        limpar_nome_do_agente.clear()

        # //*[@id="formulario:dados"]/table/tbody/tr/td/div/table/tbody/tr[11]/td[3]/input[1]
                                                            
    except Exception as e:
        pass

    # Iterando sobre os dados
for indice, row in dados.iterrows():
    imei = str(row['IMEI'])
    instalacao = str(row['N√öMERO DE INSTALA√á√ÉO'])
    #coletor = row['COLETOR']
    #controle = str(row['CONTROLE'])
    fabricante = row['FABRICANTE']
    modelo = row['MODELO']
    nome = row['NOME']
    #nso = str(row['SO'])

    print(f'Cadastrando {indice} - {imei}')
    cadastrar_aparelho(imei, instalacao, fabricante, modelo, nome)

```
